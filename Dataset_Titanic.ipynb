```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv('https://raw.githubusercontent.com/muliasulistiyono/ST152-Big-Data-Predictive-Analysis/main/Materi%20Praktikum/Pertemuan%203/titanic%20-%20titanic.csv')

# NaN Values Converter :)
df_clean = df.dropna(subset=['Age', 'Fare'])

# Calculate Coefficient
correlation, p_value = pearsonr(df_clean['Age'], df_clean['Fare'])

print(f"The correlation coefficient between Age and Fare is: {correlation:.2f}")

# Create Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['Age'], df_clean['Fare'])
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```
